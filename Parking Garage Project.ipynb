# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
-leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
# Your Code will go here
from IPython.display import clear_output

class ParkingGarage():
    """
    The ParkingGarage class will have tickets and  
    parking spaces.
    
    Attributes for the class:
    - tickets: expected to be a list
    - parkingSpaces: expected to be a list
    - currentTicket: expected to be a dictionary
    
    Methods for the class:
    - takeTicket: decreases tickets and parkingSpaces by 1
    - payForParking: takes user input for payment on current ticket
    - leaveGarage: requests payment or 
    """
    
    def __init__(self):
        self.tickets = list(range(0,20))
        self.parkingSpaces = list(range(0,20))
        self.currentTicket = {"lastTicket": 0, "lastSpace": 0}
        
    def takeTicket(self):
        # This should decrease the amount of tickets available by 1
        # This should decrease the amount of parkingSpaces available by 1
        if self.currentTicket["lastTicket"] == 0:
            self.currentTicket["ticket"] = self.tickets.pop()
            self.currentTicket["parkingSpace"] = self.parkingSpaces.pop()
            self.currentTicket["paid"] = False
        else:
            self.currentTicket["lastTicket"] = self.currentTicket["ticket"]
            self.currentTicket["lastSpace"] = self.currentTicket["parkingSpace"]
            self.currentTicket["ticket"] = self.tickets.pop()
            self.currentTicket["parkingSpace"] = self.parkingSpaces.pop()
            self.currentTicket["paid"] = False
        space = self.currentTicket["parkingSpace"]
        print(f"Your parking space is {space}")

        
    def payForParking(self):
        # Display an input that waits for an amount from the user and store it in a variable
        # If the payment variable is not empty then -> display a message to the user that their ticket has been paid and they have 15mins to leave
        # This should update the "currentTicket" dictionary key "paid" to True
        payment = input("How would you like to pay?")
        if payment != "":
            print("Payment recieved")
            self.currentTicket["paid"] = True
        else:
            self.currentTicket["paid"] = False
            
    def leaveGarage(self):
        # -If the ticket has been paid, display a message of "Thank You, have a nice day"
        # -If the ticket has not been paid, display an input prompt for payment
        # -Once paid, display message "Thank you, have a nice day!"
        # -Update parkingSpaces list to increase by 1
        # -Update tickets list to increase by 1
        if self.currentTicket["paid"] == True:
            print("Thank you, have a nice day")
        else:
            print("Payment required")
            while(self.currentTicket["paid"] == False):
                self.payForParking()
            print("Thank you for parking with us, have a nice day")
        self.parkingSpaces.append(len(self.parkingSpaces)+1)
        self.tickets.append(len(self.tickets)+1)
        

def lets_park():
    """
    Make and navigate a ParkingGarage menu.
    
    Takes user input to navigate a parking garage menu, the
    options are take ticket, pay for parking, or leave garage. 
    - Take Ticket: 
    - Pay for Parking: 
    - Leave Garage: 
    """
    print("Welcome to the Parking Garage")
    garage = ParkingGarage()
    while True:
        menu_choice = input("Do you want to:\n- Take Ticket\n- Pay for Parking\n- Leave Garage\n")
        if menu_choice.lower() == 'take ticket':
            clear_output()
            garage.takeTicket()
        elif menu_choice.lower() == 'pay for parking':
            garage.payForParking()
        elif menu_choice.lower() == 'leave garage':
            garage.leaveGarage()
            break
        else:
            clear_output()
            print("Option not recognized, try again!")

lets_park()
```
